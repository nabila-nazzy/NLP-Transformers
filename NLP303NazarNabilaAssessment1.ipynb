**Summerising an Article using a transformer**
We are using the Facebook bart arge CNN transformer to perform this task.

````python
from transformers import pipeline

summarizer = pipeline("summarization", model = "facebook/bart-large-cnn")
# First we are initialising the model and testing to see if it works without an interface.

def predict(prompt):
    completion = summarizer(prompt)
    return completion
````

# Text Generator
This section is a text generator which uses the GPT2 model to do the task.

````python
from transformers import pipeline
# Importing the pipeline needed for the task.

# First we need to load the model using the pipeline function!
generator = pipeline('text-generation', model = 'gpt2', max_length = 500)

# Then we need to define a text generation function.
# This function will take the input prompt and return a text generation result based on the input.
def ai_text(prompt):
    completion = generator(prompt)[0]["generated_text"]
    return completion
````

````python
ai_text("Guts is a beserk character") # We are testing to see if the generation function works.

# Notes: I've set the maximum length of generation to be 500 words.
````

````python
# We are now creating an interfact for the text generation model we have just put in our function.

import gradio as gr

examples = ["Guts is a beserk character"]

             # Loading our function which has the model here.
gr.Interface(fn=ai_text, title = "Text Generation", theme = "peach", inputs="text", outputs="text", examples=examples).launch()
````

# Translation Model

# Masked Language Modelling
aka text prediction will be done in this section.
Will be using the distilbert-base-uncased model for this task.

````python
from transformers import pipeline

# loading the model to see if it works.
unmasker = pipeline('fill-mask', model='distilbert-base-uncased')
unmasker("Hello I'm a [MASK] model.")
````

## Masked Language Model
Deployed with Gradio

````python
# Defining a function that outputs the token string, score, and sequences seperately for the UI to display.
# The UI will only desplay the FIRST RESULT of the unmasker starting with the number 0 on the list.
def unmasker1(text):
    results = unmasker(text)
    return results[0]["token_str"], results[0]["score"], results[0]["sequence"]
````

````python
import gradio as gr

# NOTE:
# Must include the word [MASK] for the masking model to work its magic if you want to write your own prompt.
examples = [
    ["The goal of life is [MASK]."],
    ["What can I [MASK]?"],
    ["Hello I'm a [MASK] model."]
]

# The UI is being generated here.
gradio_ui = gr.Interface(
    fn=unmasker1,
    title="Masked Language Modelling", examples=examples,
    description="Enter some text and see if the Distilbert model can MASK it",
    inputs=gr.inputs.Textbox(lines=1, label="Paste some text here"),
    outputs=[gr.outputs.Textbox(label="Label"),gr.outputs.Textbox(label="Score"),gr.outputs.Textbox(label="Sequence"),],
    )


gradio_ui.launch(share=True) # LAUNCHING THE UI yippieee


# Please read:
# You need to scroll DOWN to see the default answers once the UI loads.
````

# Chatting Bot with Microsoft DialoGPT
This section will present using a chatting bot with the help of the DialoGPT model by Microsoft. This is useful for customer service things seen in many websties.

````python
# Importing the Library for the chat bot, using microsoft dialoGPT.

from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-large")
model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-large")
````

## **Text Chatbot 1**
This is the chatbot model without the Gradio interface, the output gives you a prompt to type in.

````python
# This example was taken directly from the model card for the sake of an example test.
# https://huggingface.co/microsoft/DialoGPT-large?text=Hey+my+name+is+Julien%21+How+are+you%3F

# Chatting with the bot for 5 lines.
for step in range(5):
    # encode the new user input, add the eos_token and return a tensor in Pytorch
    new_user_input_ids = tokenizer.encode(input(">> User:") + tokenizer.eos_token, return_tensors='pt')

    # append the new user input tokens to the chat history
    bot_input_ids = torch.cat([chat_history_ids, new_user_input_ids], dim=-1) if step > 0 else new_user_input_ids

    # generated a response while limiting the total chat history to 1000 tokens, 
    chat_history_ids = model.generate(bot_input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)

    # pretty print last ouput tokens from bot
    print("DialoGPT: {}".format(tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)))
````

## **Text Chatbot 2**
This is the chatbot model hosted on gradio with the Gradio interface for input.

````python
def predict(input, history=[]):
    # tokenize the new input sentence
    new_user_input_ids = tokenizer.encode(input + tokenizer.eos_token, return_tensors='pt')

    # append the new user input tokens to the chat history
    bot_input_ids = torch.cat([torch.LongTensor(history), new_user_input_ids], dim=-1)

    # generate a response 
    history = model.generate(bot_input_ids, max_length=4000, pad_token_id=tokenizer.eos_token_id).tolist()

    # convert the tokens to text, and then split the responses into lines
    response = tokenizer.decode(history[0]).split("<|endoftext|>")
    #print('decoded_response-->>'+str(response))
    response = [(response[i], response[i+1]) for i in range(0, len(response)-1, 2)]  # convert to tuples of list
    #print('response-->>'+str(response))
    return response, history
````

````python
import torch
import gradio as gr

description = "This is a chatbot application based on the DialoGPT model of Microsoft."
title = "Chatbot <3"
examples = [["Hello! How are you?"]]
gr.Interface(fn=predict,
             title=title,
             description=description,
             examples=examples,
             inputs=["text", "state"],
             outputs=["chatbot", "state"]).launch(share=True)
````
